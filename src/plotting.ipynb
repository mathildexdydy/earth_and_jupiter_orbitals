# Sun–Earth–Jupiter Simulation Plots

This notebook loads the CSV files produced by the C++ simulation and plots:

- Earth and Jupiter orbits
- Total energy vs time

Expected files in the same folder:

```
orbit_3600.csv
orbit_86400.csv
orbit_604800.csv
energy_3600.csv
energy_86400.csv
energy_604800.csv
```

````python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

data_path = Path('.')
````

## Plot Orbits

````python
timesteps = [3600, 86400, 604800]

for dt in timesteps:
    file = data_path / f'orbit_{dt}.csv'
    df = pd.read_csv(file)

    plt.figure()
    plt.plot(df['xe'], df['ye'], label='Earth')
    plt.plot(df['xj'], df['yj'], label='Jupiter')
    plt.scatter([0],[0], label='Sun')

    plt.xlabel('x (m)')
    plt.ylabel('y (m)')
    plt.title(f'Orbits (dt = {dt} s)')
    plt.legend()
    plt.axis('equal')
    plt.show()
````

## Plot Total Energy

````python
for dt in timesteps:
    file = data_path / f'energy_{dt}.csv'
    df = pd.read_csv(file)

    plt.figure()
    plt.plot(df['time_days'], df['E'])

    plt.xlabel('Time (days)')
    plt.ylabel('Total Energy (J)')
    plt.title(f'Energy Conservation (dt = {dt} s)')
    plt.show()
````
